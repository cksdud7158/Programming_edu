```python
# 가장 팁이 높은 데이타 상위 5개를 출력 
tipdf.sort_values('tip', ascending=False).head()
```

```python
def get_tippct(df):
    return round(df['tip']/df['total_bill']*100,1)
```

```python
# 1. tip_pct 를 구하자...tip_pct를 새로운 컬럼으로 추가 
# 2. get_tippct 함수로 구한값을 컬럼의 값으로 지정
tipdf['tip_pct'] = get_tippct(tipdf)
tipdf.sort_values(['total_bill','tip_pct'],ascending=[False,False]).head(7)
```

```python
tipdf.sort_values(['total_bill','tip_pct'],ascending=[False,False]).head(7).plot(kind='bar')
plt.show()
```

```python
# Male의 정보를 따로 분리해서 tipdf_man이라는 데이타 프레임을 만들어라
tipdf_man = tipdf[tipdf['sex']=='Male']
tipdf_man.shape
tipdf_man.head(10)
```

```python
# Feale의 정보를 따로 분리해서 tipdf_woman이라는 데이타 프레임을 만들어라
tipdf['sex'].unique()
# tipdf_woman = tipdf[tipdf['sex'].isin('Female')] #[]형태로만 가능하다.
tipdf_woman = tipdf[tipdf['sex'].isin(['Female'])]
tipdf_woman.head(10)
```

```python
# tipdf_man, tipdf_woman 두개의 데이타 프레임을 병합해서
# tip_all 이라는 데이타 프레임 생성
tip_all = pd.concat([tipdf_man,tipdf_woman],keys=['Male','Female'])
tip_all
```
