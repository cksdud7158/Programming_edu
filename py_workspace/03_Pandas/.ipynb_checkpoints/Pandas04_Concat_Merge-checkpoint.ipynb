```python
import numpy as np
import pandas as pd 

from pandas import Series, DataFrame
import matplotlib.pyplot as plt


from numpy import nan as NA
```

```python
'''
    서로 다른 데이타를 하나로 합치는 작업 
    1) Concat
        단순히 하나의 데이타 프레임에 다른 데이타 프레임을 연속적으로 붙이는 방법 
        이 경우에는 두개의 데이타 프레임의 인덱스와 컬럼이 서로 동일한 경우가 대부분
        기본이 위, 아래로 연결
        Axis 를 수정하면 좌우 연결 가능 
        
        outer join 이 기본 방식이다. 
        key를 이용한 concate이 가능하다. 
        
    2) Merge
    
'''
```

```python
df1 = DataFrame({
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']
})
df1
```

```python
df2 = DataFrame({
    'A':['A4','A5','A6','A7'],
    'B':['B4','B5','B6','B7'],
    'C':['C4','C5','C6','C7'],
    'D':['D4','D5','D6','D7']
})
```

```python
'''
    df1, df2 의 데이타를 보면 상하로 병합해야되는 데이타라는걸 알수 있다. 
    인덱스가 0,1,2,3,0,1,2,3
    동일한 값이 반복되어져있다. 
'''
result = pd.concat([df1,df2], ignore_index=True)
result
```

```python
# 좌우 연결도 가능하다. 
result = pd.concat([df1,df2], axis=1)
result
```

```python
'''
    데이타를 합칠떄 어느 그룹에서 가져왔는지 구분이 가능 
    또한 keys 속성을 통해서 데이타의 그룹핑이 된다. 
'''
result = pd.concat([df1,df2], keys=['ClassA', 'ClassB'])
result
```

```python
df3 = DataFrame({
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
    'C':['C0','C1','C2','C3']
})
df3
```

```python
df4 = DataFrame({
    'A':['A4','A5','A6','A7'],
    'B':['B4','B5','B6','B7'],
    'C':['C4','C5','C6','C7'],
    'D':['D4','D5','D6','D7']
})
df4
```

```python
# df3, df4 는 행과 열이 동시에 일치하지않는다. 병합시 문제 발생
# df3 에는 D 열이 없기 떄문에 Nan으로 채워진다. 
result2 = pd.concat([df3,df4], ignore_index=True)
result2
# result2.fillna(method='ffill') # 첫번쨰 컬럼이 없어서 안채워짐 
result2.fillna('A') # 첫번쨰 컬럼이 없어서 안채워짐 
```

```python
result3 = pd.concat([df3,df4], join='outer')
result3

# join='inner' 방식으로 데이타프레임들을 연결하면 두 데이타에 모두 존재하는 행렬 값만 가져온다.
result3 = pd.concat([df3,df4], join='inner')
result3
```
