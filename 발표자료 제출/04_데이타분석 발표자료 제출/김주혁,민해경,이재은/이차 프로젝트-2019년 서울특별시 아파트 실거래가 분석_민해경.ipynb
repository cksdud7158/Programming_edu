## 2. 실거래가 데이터 가공

```python
apt = pd.read_csv('data/2019년 서울 아파트 실거래가.csv', encoding="cp949")
apt.head()
```

```python
apt.info()
```

```python
# 거래금액 컬럼의 콤마(,) 제거 후 데이터 타입 변경..int64로
# lst = apt['거래금액'][1].split(',')
# lst[0] = lst[0].replace(' ', '')
# a = int("".join(lst))
# type(a)

def convert_int(x):
    lst = x.split(',')
    lst[0] = lst[0].replace(' ', '')
    return int("".join(lst))
```

```python
apt['거래금액'] = apt['거래금액'].apply(convert_int)
apt
```

```python
# 전용면적(제곱미터)를 평으로 변환(나누기 3.306)하여 전용면적(평) 컬럼 생성 후 소숫점 둘쨰자리까지 표시합니다.
apt['전용면적(평)'] = round(apt['전용면적']/3.306, 2)
apt
```

```python
# 전용면적을 범주화 하기 위해 전용면적(평) 컬럼을 정수형으로 바꾼 새로운 컬럼을 생성합니다.
# 컬럼명 전용면적2(평)
from math import floor
apt['전용면적(평)2'] = apt['전용면적(평)'].map(floor)
apt
```

```python
a = 29//5
b, c = a*5, (a+1)*5
'(' + str(b) + ', ' + str(c) + ']'
```

```python
# 전용면적을 범주화 합니다.
# 컬럼명은 전용면적(평)3

def categorized_area(x):
    # x에서 가까운 5의 배수를 찾는다
    a = x // 5
    b, c = a*5, (a+1)*5
    return '(' + str(b) + ', ' + str(c) + ']'


def categorized_area1(x):
    # x에서 가까운 5의 배수를 찾는다
    if x < 10:
        b, c = 10, 20
    a = x // 10
    b, c = a*10, (a+1)*10
    return '(' + str(b) + ', ' + str(c) + ']'
```

```python
apt['전용면적(평)3'] = apt['전용면적(평)2'].map(categorized_area1)
apt
```

```python
apt['전용면적(평)3'] = apt['전용면적(평)3'].astype('category')
```

```python
apt.info()
```

```python
#### 범주화된 전용면적의 고유값을 확인합니다.
apt['전용면적(평)3'].unique()
```

## 3. 데이타 프레임 병합

```python
# 법정동코드(locCode)와 아파트 실거래가(apt) 합치기
# 강일동 : 11740

locCode['지역코드'] = locCode['법정동코드'].map(lambda x: int(str(x)[:5]))
locCode['법정동'] =locCode['읍면동명']

locCode1 = locCode.drop_duplicates('시군구명')[['시군구명', '지역코드']]
locCode1
```

```python
# innner join
df = pd.merge(locCode1, apt, on='지역코드')
df
```

```python
df['평단가'] = round(df['거래금액']/df['전용면적(평)2'], 0)
df.head()
```
