```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
```

```python
# 상반기 주유소 판매 데이터 불러오기
first = pd.read_csv('../data/2019년 상반기 주유소 판매가격.csv', encoding='cp949')
first.head()
```

```python
# 하반기 주유소 판매 데이터 불러오기
second = pd.read_csv('../data/2019년 하반기 주유소 판매가격.csv', encoding='cp949')
second.head()
```

```python
# 상반기 판매가격 데이터 프레임 결측치 조회
first.isnull().sum()
```

```python
# 하반기 판매가격 데이터 프레임 결측치 조회
second.isnull().sum()
```

```python
# 상반기와 하반기 데이터를 상하로 결합
oil_station = pd.concat([first,second]).drop_duplicates()
oil_station
```

```python
# 데이터 프레임 정보 조회
oil_station.info()
```

```python
# 상표 컬럼의 고유값 조회
oil_station['상표'].unique()
```

```python
# 지역 컬럼의 고유값 조회
oil_station['지역'].unique()
```

```python
# 지역 컬럼의 값을 이용하여 시와 구를 분리
oil_station[['시', '구']] = oil_station['지역'].str.split(' ', n=1, expand=True)
oil_station.head()
```

```python
# 시 컬럼의 서울을 서울특별시로 변환
oil_station['시'] = oil_station['시']+"특별시"
oil_station.head()
```

```python
# 기간 컬럼을 datetime 형식으로 변경
oil_station['기간'] = oil_station['기간'].astype(str)
oil_station['기간'] = pd.to_datetime(oil_station['기간'])
oil_station['기간']
```

```python
# 기간 컬럼에서 dt 타입을 이용하여 년, 월, 일, 요일 컬럼 생성
oil_station['년'] = oil_station['기간'].dt.year
oil_station['월'] = oil_station['기간'].dt.month
oil_station['일'] = oil_station['기간'].dt.day
oil_station['요일'] = oil_station['기간'].dt.day_name()
oil_station.head()
```

```python
# 휘발유 가격이 비싼 5개 판매가격을 출력하세요.
oil_station.sort_values(by='휘발유', ascending=False).head()
```

```python
# 휘발유 가격이 저렴한 5개 판매가격을 출력하세요. (단 휘발유가 0원인 데이터는 제외)
oil_station[oil_station['휘발유']!=0].sort_values(by='휘발유').head()
```

```python
# 상표별 평균 휘발유 가격을 계산하세요.
oil_station.rename(columns={'휘발유':'mean'}).pivot_table('mean', index='상표')
```

```python
# 상표별 평균 휘발유 가격을 시각화하세요.
oil_station.rename(columns={'휘발유':'mean'}).pivot_table('mean', index='상표').plot(kind='bar')
```

```python
# 상표와 셀프여부별 평균 휘발유 가격을 계산하세요.
oil_station.rename(columns={'휘발유':'mean'}).pivot_table('mean', index=['상표','셀프여부'])
```

```python
# 상표와 셀프여부별 평균 휘발유 가격을 시각화하세요.
oil_station.rename(columns={'휘발유':'mean'}).pivot_table('mean', index=['상표','셀프여부']).plot(kind='bar')
```

```python
# 셀프여부에 따라 휘발유 가격을 boxplot으로 시각화해보세요.
# 휘발유가 0인 데이터는 제외하고 시각화합니다.
plt.figure(figsize=(12,7))
sns.boxplot(x='셀프여부', y='휘발유', data=oil_station[oil_station['휘발유']!=0])
plt.show()
```

```python
# 휘발유 가격과 일반 주유소의 가격을 boxplot을 이용하여 시각화해보세요.
# 휘발유가 0인 데이터는 제외하고 시각화합니다.
plt.figure(figsize=(12,7))
sns.boxplot(x='상표', y='휘발유', hue='셀프여부', data=oil_station[oil_station['휘발유']!=0])
plt.show()
```

```python
# 지역별 평균 휘발유 가격을 계산하세요
oil_price_by_area = oil_station.rename(columns={'휘발유':'mean'}).pivot_table('mean', 
                                        index='구').sort_values(by='mean', ascending=False)
oil_price_by_area 
```

```python
# 지역별 평균 휘발유 가격을 시각화하세요.
oil_price_by_area.plot(kind='bar')
```

```python
# 월별 평균 휘발유 가격을 계산하세요
oil_price_by_month = oil_station.rename(columns={'휘발유':'mean'}).pivot_table('mean', 
                                        index='월')
oil_price_by_month
```

```python
# 월별 평균 휘발유 가격을 시각화 하세요.
oil_price_by_month.plot(kind='bar')
```

```python
# pivot_table 함수를 이용하여 구와 월 컬럼을 이용하여 평균 휘발유 가격을 계산하세요
oil_price_by_area_month = oil_station.pivot_table('휘발유', index='월', columns='구')
oil_price_by_area_month
```

```python
# pivot_table 함수를 이용하여 구와 월 컬럼을 이용하여 평균 휘발유 가격을 시각화 하세요 (heatmap 이용)
plt.figure(figsize=(10,8))
sns.heatmap(oil_price_by_area_month) 
plt.show()
```

```python
# 시간에 따른 상표별 휘발유 가격변화
oil_price_by_time = oil_station.pivot_table('휘발유', index=['기간','상표'])
oil_price_by_time = oil_price_by_time.reset_index()
oil_price_by_time
```

```python
oil_price_by_time = oil_station.pivot_table('휘발유', index='기간', columns='상표')
oil_price_by_time

plt.style.use(['bmh'])
plt.rcParams["figure.figsize"] = (10, 7)
oil_price_by_time.plot()
plt.title('시간에 따른 상표별 휘발유 가격 변화')
plt.xlabel('기간')
plt.ylabel('가격')
plt.show()
```
